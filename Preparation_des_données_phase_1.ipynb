```python
import pandas as pd
import numpy as np
```

```python
source="C:/Users/HP/Desktop/Commitment_to_health_by-country_2007-2008.xlsx"
df = pd.read_excel(source)
df.head(12)
```

```python
df = pd.read_excel(source, header=6)
df.head(12)
```

```python
df = df.dropna(how="all", axis=1)
df.head(12)
```

```python
df = df.dropna(how="all", axis=0).reset_index(drop=True)
df.head(12)
```

```python
colonnes=[]
for c in df.columns:
     colonnes.append([str(i) for i in df[c][:4] if pd.notnull(i)])           
colonnes
```

```python
colonnes=[
    "HDI_rank",
    "Countries",
    "Health_expenditure_Public(%)",
    "Reference1",
    "Health_expenditure_Private(%)",
    "Reference2",
    "Health_expenditure_Per capita(PPA)",
    "One-year-olds_fully_immunized_Against_tuberculosis(%)2005",
    "Reference3",
    "One-year-olds_fully_immunized_Against_measles(%)2005",
    "Reference4",
    "Child_diarr_receiv_oral_rehydration(% under age 5)",
    "Reference5",
    "Contraceptive_prevalence_rate(%)",
    "Reference6",
    "Births_attended_by_skilled_health_personnel(%)",
    "Reference7",
    "Physicians_per_100000_people",
    "Reference8"
]
```

```python
df.columns=colonnes
df.head(8)
```

```python
df=df.drop(np.arange(4),axis=0).reset_index(drop=True)
df.head()
```

```python
df.tail(23)
```

```python
df=df.drop(np.arange(180,len(df)),axis=0).reset_index(drop=True)
df.tail()
```

```python
hdi_categories = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "LOW HUMAN DEVELOPMENT"]
df=df.drop(df[df["HDI_rank"].isin(hdi_categories)].index,axis=0).reset_index(drop=True)
df.head()
```

```python
df.loc[:, "HDI_category"] = np.where(df["HDI_rank"] > 155,
                                     "LOW HUMAN DEVELOPMENT",
                                     np.where(df["HDI_rank"] < 71,
                                              "HIGH HUMAN DEVELOPMENT",
                                              "MEDIUM HUMAN DEVELOPMENT"
                                              
                                             )
                                    )
df.sort_values(by="HDI_rank", inplace=True)
df.head()
```
